```python
df=df.dropna(subset=['dob'])
df
```

```python
df.loc[:,'age']=df['age'].fillna(df['age'].median())
df.loc[:,'age_missing']=df['age'].isna().astype(int)
df=df.drop_duplicates(subset=['dob'])
```

```python
df[['dob','age']].head(10)
```

```python
df=df.drop(columns=['age_missing'])
df
```

```python
df=df.copy()
```

```python
df['startdate']=pd.to_datetime(df['startdate'],errors='coerce')
df['exitdate']=pd.to_datetime(df['exitdate'],errors='coerce')
exit_filled=df['exitdate'].fillna(pd.Timestamp.today())
```

```python
tenure_years=(exit_filled-df['startdate']).dt.days/365
df['tenure_group']=pd.cut(tenure_years,bins=[0,1,3,5,10,50],labels=['0-1','1-3','3-5','5-10','10+'])
df[['startdate','exitdate','tenure_group']].head(10)
df
```

```python
df['exitdate'] = pd.to_datetime(df['exitdate'], format='%d-%b-%y', errors='coerce')
df['attrition'] = df['exitdate'].notna().astype(int)
```

```python
df['performance_level'] = df['performance_score'].map({
    'Exceeds': 'High',
    'Fully Meets': 'Medium',
    'Needs Improvement': 'Low',
    'PIP': 'Low'
})
```
